##Loading the Dataset

```python
import pandas as pd

df = pd.read_csv("housing (4).csv")
df.head()
```

##PART A — Method Chaining

```python
summary = (
    df[(df["price"] > 250000) & (df["size"] > 1000)]
    .assign(price_per_sqft=lambda x: x["price"] / x["size"])
    .groupby("neighborhood")
    .agg(
        mean_price_per_sqft=("price_per_sqft", "mean"),
        median_price_per_sqft=("price_per_sqft", "median"),
        count_homes=("price_per_sqft", "count")
    )
    .sort_values(by="mean_price_per_sqft", ascending=False)
)

summary
```

##PART B — dplyr Version

```python
summary_dplyr_equivalent = (
    df[(df["price"] > 250000) & (df["size"] > 1000)]
    .assign(price_per_sqft=lambda x: x["price"] / x["size"])
    .groupby("neighborhood")
    .agg(
        mean_price_per_sqft=("price_per_sqft", "mean"),
        median_price_per_sqft=("price_per_sqft", "median"),
        count_homes=("price_per_sqft", "count")
    )
    .sort_values(by="mean_price_per_sqft", ascending=False)
)

summary_dplyr_equivalent
```

##Reflection

I find pandas method chaining slightly more complex than dplyr pipelines because the syntax requires more attention to detail. However, both approaches follow a similar structure where data is filtered, transformed, grouped, and summarized step by step. The similarity is that both use a pipeline approach to make transformations readable. The difference is that dplyr feels more intuitive, while pandas requires more explicit syntax.
